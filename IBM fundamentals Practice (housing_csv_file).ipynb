Building Visualization to understand the dataframe and its features

```python
pyplot.bar(housing_df['bedrooms'], housing_df['price'])
pyplot.xlabel('Bedrooms')
pyplot.ylabel('Price in Millions')
pyplot.title('Bedrooms vs. Price in Millions')
pyplot.show() #optional
#exploration part of data
```

```python
pyplot.barh(housing_df['bedrooms'], housing_df['price'])
pyplot.ylabel('Bedrooms')
pyplot.xlabel('Price in Millions')
pyplot.title('Bedrooms vs. Price in Millions')
pyplot.show()
```

```python
# prompt: pie chart showing furnishing status on housing_df data frame

import matplotlib.pyplot as plt
# Pie chart showing furnishing status on housing_df data frame
furnishing_status = housing_df['furnishingstatus'].value_counts()
labels = furnishing_status.index.to_list()
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.pie(furnishing_status, autopct="%1.1f%%", labels=labels, colors=colors, startangle=90, wedgeprops={'edgecolor': 'white', 'linewidth': 1})
plt.title("Furnishing Status")
plt.show()
```

```python
pyplot.scatter(housing_df['bathrooms'], housing_df['price'])
pyplot.xlabel('Bathrooms')
pyplot.ylabel('Price in Millions')
pyplot.title('Bathrooms vs. Price in Millions')
pyplot.show()
```

```python
housing_df.plot(x='price', y='area', kind ='line')
pyplot.ylabel('Area')
pyplot.xlabel('Price in Millions')
pyplot.title('Price in Millions vs. Area')
pyplot.show()
```
